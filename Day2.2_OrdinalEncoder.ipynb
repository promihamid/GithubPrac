**Ordinal Encoding: Ordinal Encoding is a technique used to convert categorical data (data with names or labels) into numbers, when the categories have a meaningful order.<br> Use ordinal encoding when your data has levels or rankings.<br>
Examples:
Education: High School < Bachelor < Master < PhD<br>
Size: Small < Medium < Large<br><br>To solve ordinal encoding, replace ordered category values with meaningful numbers using either `OrdinalEncoder` from scikit-learn or the `map()` function in pandas**

```
import pandas as pd
```

```
dataset = pd.read_csv('/content/loan.csv')
```

```
dataset.head(3)
```

```
dataset['Property_Area'].unique()
```

```
dataset.isnull().sum()
```

**NO null value so no need to fill it**

```
en_data_ord=[['Rural', 'Semiurban','Urban']]
```

```
from sklearn.preprocessing import OrdinalEncoder
```

```
oen = OrdinalEncoder(categories=en_data_ord)
```

```
dataset['Property_Area']=oen.fit_transform(dataset[['Property_Area']])
```
