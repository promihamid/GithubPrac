```python
import pandas as pd
```

```python
dataset = pd.read_csv("/content/loan.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

**Filling missing value**

```python
dataset["Gender"].fillna(dataset["Gender"].mode()[0], inplace=True)
dataset["Married"].fillna(dataset["Married"].mode()[0], inplace=True)
dataset["Self_Employed"].fillna(dataset["Self_Employed"].mode()[0],inplace=True)
```

**One hot encoding**

```python
en_data = dataset[["Gender","Married","Self_Employed"]]
```

**One hot encoding using get_dummies() method**

```python
pd.get_dummies(en_data) #drop_first for avoiding redundancy, get_dummies() used for one-hot encoding
```

**ekhane value still string e ache but we want it in numerical format. So use OneHotEncoder from sklearn <br><br>
One hot encoding using sklearn**

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop="first")
ar = ohe.fit_transform(en_data).toarray() # fit_transform--> 1st e data ke dekhe then sckitlearn er algo apply kore, then transform kore numerical e
```

use toarray() to convert the data into array. but we want out DATA AS DATAFRAME FORMAT NOT IN ARRAY FORMAT. so use dataframe() method

```python
pd.DataFrame(ar,columns=["Gender_Male","Married_Yes","Self_Employed_Yes"])
```
