<h1><b><font color='red'>Handling duplicate data <br>
<h5><b><font color='black'>Duplicate data means the same row appears more than once in your dataset. This can cause wrong results. You can find duplicates by dataset.duplicated().sum(). To remove them, use dataset.drop_duplicates(inplace=True) which keeps the first copy.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data = {
    'Name': ['Alice', 'Bob', 'Alice', 'David', 'Bob'],
    'Age': [25, 30, 25, 40, 30]
}
```

```
df = pd.DataFrame(data)
```

<b>`df = pd.DataFrame(data)` creates a table (DataFrame) from the dictionary `data` where keys are column names and values are the column data.


```
df
```

```
 df.duplicated().sum()
```

```
df.drop_duplicates(inplace=True)
```
