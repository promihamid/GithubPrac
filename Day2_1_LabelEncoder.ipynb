**Label Encoding <br><br>Label Encoding:Label Encoding is a method to convert text (categorical) values into numbers. NOMINAL DATA er upor hoy.<br>Nominal data has no order (e.g., `Color = Red, Blue, Green`), while Ordinal data has a clear order (e.g., `Size = Small, Medium, Large**

**fit --> model ke train kore**

```
import pandas as pd
```

```
dataset=pd.read_csv("loan.csv")
dataset.head(3)
```

```
from sklearn.preprocessing import LabelEncoder
```

```
dataset['Property_Area'].unique() #optional work, used to check the number of unique values present in the Property_area
```

```
le = LabelEncoder()
le.fit(dataset['Property_Area'])
```

```
dataset['Property_Area']=le.transform(dataset['Property_Area'])
```
