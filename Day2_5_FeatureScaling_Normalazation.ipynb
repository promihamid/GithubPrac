**<h1><font color='red'>Feature Scaling<br><h6><font color='black'>It's like making sure everyone in a race starts from the same line. In machine learning, “features” are your dataset’s columns. If one column has big numbers (like salary) and another has small numbers (like age), the bigger one can unfairly influence the model. So, we "scale" them to bring everything to the same level.**

<h3><b><font color='red'>Standardization

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
dataset = pd.read_csv('/content/loan.csv')
```

```
dataset.isnull().sum()
```

```
dataset.describe()
```

```
sns.displot(dataset['CoapplicantIncome'])
plt.show()
```

<b><h1><font color='red'>Min-MaxScaler

```
from sklearn.preprocessing import MinMaxScaler
```

```
ms = MinMaxScaler()
ms.fit(dataset[['CoapplicantIncome']])
```

**Line 1: ms = MinMaxScaler()
Creates a MinMaxScaler object named ms.
This scaler will later be used to scale values between 0 and 1.<br>
Line 2: ms.fit(dataset[['CoapplicantIncome']])
It learns the minimum and maximum values of the 'CoapplicantIncome' column.
It does not scale yet — it only stores the min and max for later use when you call .transform().**

```
dataset['CoapplicantIncome_min'] = ms.transform(dataset[['CoapplicantIncome']])
```

<b>This line **scales** the `'CoapplicantIncome'` values to the **0–1 range** using the `ms` scaler,
and saves the scaled result in a **new column** called `'CoapplicantIncome_min'`.


```
plt.figure(figsize=(10, 7))

# First subplot (Before scaling)
plt.subplot(1, 2, 1)
plt.title('Before')
sns.histplot(dataset['CoapplicantIncome'], kde=True)  # Use histplot instead of displot

# Second subplot (After scaling)
plt.subplot(1, 2, 2)
plt.title('After')
sns.histplot(dataset['CoapplicantIncome_min'], kde=True)

#plt.tight_layout()
plt.show()
```
